BELOW IS UNPACKING OF A TUPLE BY INDEXES
This process is called unpacking.

Why "unpacking"?

*Because you're taking one object (the tuple) and unpacking its contents into multiple variables.*


```python
tuple = (1,2,3)
x = tuple[0]
y = tuple[1]
z = tuple[2]
print(x)
print(y)
print(z)
```

Below is the unpacking of the tuple


```python
tuple = ('Student','BSCS-2006-10023' )
x,y = tuple
print(x)
print(y)
```
