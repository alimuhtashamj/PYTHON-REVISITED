```python
scores = {
    "Ali": 90,
    "Ahmed": 80,
    "Sara": 95
}
print(scores.keys())
```

```python
print(scores.values())
for value in scores.values():
    print(value)
```

```python
print(scores.items()) #gives back results in tuple
for key, value in scores.items():
    print(key, value) #unpacking of tuple
```

```python
for score in scores.values():
    print(score + 5)
```

```python
employees = {
    101: {
        "name": "Ali",
        "department": "AI"
    },
    102: {
        "name": "Ahmed",
        "department": "Backend"
    }
}
for emp_id, info in employees.items():
    print(emp_id, info)
```
