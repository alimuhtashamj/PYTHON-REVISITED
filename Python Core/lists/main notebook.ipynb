A sports team is standing in order of their player numbers. The coach wants to change the strategy by splitting the team in half and reversing the order of the first half.

`````python
team_numbers = [10, 14, 22, 35, 40, 51, 62, 77]
mid = len(team_numbers)//2
first_half = team_numbers[:mid]
second_half = team_numbers[mid:]
first_half.reverse()
team_numbers = first_half + second_half
print(team_numbers)
`````
