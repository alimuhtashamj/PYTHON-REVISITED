```python
string = 'abcdefghijklmnop'
# [::-1] works because leaving start/stop blank tells Python:
# "Grab everything." The -1 step tells it: "Read right-to-left."
# If we used [-1:0:1], the '1' forces it to read left-to-right,
# but nothing exists to the right of the last character (-1)!
# Plus, an explicit stop of '0' would skip the first letter anyway.

# Your starting point (-1) and stopping point (0) required moving backward.
# Your step (1) forced Python to move forward.
reverse = string[::-1]
print(reverse)
```
