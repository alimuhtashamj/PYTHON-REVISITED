
## Problem 1

Write a function:

```python
calculate_total(cart)
```

Given:

```python
cart = [
    {"name": "Laptop", "price": 1000},
    {"name": "Mouse", "price": 50},
    {"name": "Keyboard", "price": 80}
]
```
Return the total price of all items.
Example:

```python
calculate_total(cart)
```
Expected output:

```python
1130
```



````python
cart = [
      {"name": "Laptop", "price": 1000},
      {"name": "Mouse", "price": 50},
      {"name": "Keyboard", "price": 80}
  ] 

def calculate_price(cart ):

  total = 0
  for i in cart:
    total += i['price']
  return total
calculate_price(cart)

 
````
