
## Problem 1

Write a function:

```python
calculate_total(cart)
```

Given:

```python
cart = [
    {"name": "Laptop", "price": 1000},
    {"name": "Mouse", "price": 50},
    {"name": "Keyboard", "price": 80}
]
```
Return the total price of all items.
Example:

```python
calculate_total(cart)
```
Expected output:

```python
1130
```



````python
cart = [
      {"name": "Laptop", "price": 1000},
      {"name": "Mouse", "price": 50},
      {"name": "Keyboard", "price": 80}
  ] 

def calculate_price(cart ):

  total = 0
  for i in cart:
    total += i['price']
  return total
calculate_price(cart)

 
````

````python
cart = [
    {"name": "Laptop", "price": 1000},
    {"name": "Mouse", "price": 50}
]

def calculate_price(cart, discount = 10 ):
  total = 0
  remaining_percentage = 1-(discount/100)
  for product in cart:
      discounted_price = product['price']* remaining_percentage
      total += discounted_price
      
  return total
discount = 50
calculate_price(cart, discount)

 
````

````python
cart = [
    {"name": "Laptop", "price": 1000, "discount": 20},
    {"name": "Mouse", "price": 50, "discount": 5}
]

def calculate_total(cart):
    total = 0 
    for product in cart:
        price = product['price']
        discounted_price =  price * (1 - (product['discount'])/100 )
        total +=  discounted_price
    return total
calculate_total(cart)
    
````
