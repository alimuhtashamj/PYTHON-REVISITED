Create two functions:

calculate_product_price(product)
calculate_cart_total(cart)

Requirements:

Return the final price of a single product from calculate_product_price().
Return the final price of the entire cart from calculate_cart_total().
calculate_cart_total() must use calculate_product_price().

`````python
cart = [
    {"name": "Laptop", "price": 1000, "discount": 20},
    {"name": "Mouse", "price": 50, "discount": 5}
]
def calculate_product_price(product):
    discount = (1-(product['discount']/100))
    discounted_price = product['price'] * discount
    return(discounted_price )

def calculated_cart_total(cart):
    total = 0
    for product in cart:
      price = calculate_product_price(product)
      total += price
    return(total) 
calculated_cart_total(cart)
`````
