```python

def get_data():
    product_id = input('Enter product id: ')
    try:
        product_id = int(product_id)
        print("Product id is" , product_id)
    except ValueError:
        print('Enter a valid product id')
        
    product_name = input('Add product name: ')
    if product_name.isdigit():
        print('Enter valid product name')
    else:
        print('Product name is', product_name)  
         
    quantity = int(input('Type total number of product'))
    
    price = int(input('Add the product price'))
    
    process_data(product_id, product_name, quantity, price)

def process_data(product_id, product_name, quantity, price):
    print("Product ID:", product_id)
    print("Product Name:", product_name)
    print("Quantity:", quantity)
    print("Price:", price)
```
