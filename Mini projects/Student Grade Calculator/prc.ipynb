```python
student = {
    "name": "Ali",
    "marks":[10,30,40,90],
    "bonus": 5 }

def validate_fields(name,marks):
    if name is None:
        print('Student Name not Found')
        return
    if len(name) == 0:
        print('Please add the name')
    if marks is None:
        print('Please add the marks of the student')
        return
    total_subjects = len(marks)
    if total_subjects == 0:
        print('No marks are added so far')
    if total_subjects < 4:
        print('Add marks of all four subjects, You have added less than 4 subjects')
    elif total_subjects > 4:
        print('Add marks of all four subjects, You have added more than 4 subjects')   
    return
validate_fields(student['name'], student['marks'])

    
```
